The function **u(x,t)** takes in an array of domain values and iterates through them for a fixed variable **t**. In this example **D**=1.

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

def u(x, t):
    results = []
    first_term = 1/(np.sqrt(4*np.pi*t))
    for xval in x:
        second_term = np.exp((-xval**2)/(4*t))
        results.append(first_term*second_term)
    return results
z = np.linspace(-2*np.pi, 2*np.pi, 100)
t = np.arange(0.5, 6, 0.5)
for tval in t:
    plt.plot(z, u(z, tval), label=str(tval))
plt.legend()
plt.xlabel('x')
plt.ylabel('Temperature u(x, t)')
plt.show()
```
